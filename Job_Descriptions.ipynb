```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# open the first file
url = r"/home/youssef/Desktop/Jobs_Data/Data_Collecting/job_descriptions.csv"
df = pd.read_csv(url) 
df
```

```python
# Create a comprehensive summary table for all columns
summary = pd.DataFrame({
    "Column Name": df.columns,
    "Data Type": df.dtypes,
    "Non-Null Count": df.notna().sum(),
    "Null Count": df.isna().sum(),
    "Null Percentage (%)": (df.isna().mean() * 100).round(2)
})

# Step 2: Descriptive statistics
desc = df.describe(include='all').T  # Transpose so columns become rows

# Merge summary with descriptive stats
summary = summary.merge(desc, left_on="Column Name", right_index=True, how="left")

# Reorder columns but only if they exist
desired_cols = [
    "Column Name", "Data Type", "Non-Null Count", "Null Count", "Null Percentage (%)",
    "count", "unique", "top", "freq", "mean", "std", "min", "25%", "50%", "75%", "max"
]

existing_cols = [col for col in desired_cols if col in summary.columns]
summary = summary[existing_cols]

# Display the summary table
summary
```

```python
import re

# columns to process
columns = [
    "Job Description",
    "Benefits",
    "skills",
    "Responsibilities",
    "Company",
    "Company Profile"
]

# convert to string
for col in columns:
    df[col] = df[col].astype(str)

# clean normal text columns
for col in ["Job Description", "Company", "Company Profile"]:
    df[col] = (
        df[col]
        .str.strip()
        .str.replace(r'\s+', ' ', regex=True)
        .str.lower()
    )

# clean and normalize list-like columns
def clean_list_column(text):
    if pd.isna(text):
        return None
    
    text = str(text).lower()
    text = re.sub(r'\s+', ' ', text)
    text = re.sub(r'[^\w\s,]', '', text)
    
    items = [i.strip() for i in text.split(",")]
    items = list(dict.fromkeys([i for i in items if i != ""]))  # remove duplicates
    
    return ", ".join(items)

for col in ["skills", "Benefits", "Responsibilities"]:
    df[col] = df[col].apply(clean_list_column)
```

```python
df.drop(columns=["Contact", "Job Description", "Benefits", "skills", "Responsibilities"], inplace=True)
df["Job Posting Date"] = pd.to_datetime(df["Job Posting Date"], errors="coerce")
df = df.drop_duplicates(subset=["Job Id"])
df.columns = [col.title() for col in df.columns]
```

```python
import json

# Convert string to dict safely
def parse_json(x):
    try:
        return json.loads(x)
    except (json.JSONDecodeError, TypeError):
        return {}  # return empty dict if invalid

# Apply parsing to the column
df['Company Profile'] = df['Company Profile'].apply(parse_json)

# Expand the JSON into separate columns
company_profile_expanded = pd.json_normalize(df['Company Profile'])

# Concatenate the new columns with the original dataframe
df = pd.concat([df.drop(columns=['Company Profile']), company_profile_expanded], axis=1)
```
